```python
from flask import *
import xml.etree.ElementTree as ET
import shutil
import random
import threading
import os
from zipfile import ZipFile
from subprocess import Popen, TimeoutExpired, PIPE, check_call, check_output
PATH = '/content'

gl = []
is_done = True

def runcmd(cmd):
  global gl
  output = get_ipython().getoutput(cmd)
  print(output)
  gl = gl + output

def training():
  runcmd('python /content/v831_restnet18/classifier_resnet_train.py ')

def trainobj():
  runcmd('python /content/v831_yolo/train.py -d custom --cuda -v tiny_yolo_v3 -hr -ms')

#slim_yolo_v2

def genmodt():
  runcmd('python test.py -d custom -v tiny_yolo_v3 --trained_model weights/custom/tiny_yolo_v3/tiny_yolo_v3_250.pth --visual_threshold 0.3 -size 224 --export')
  runcmd('rm -rf /images')
  runcmd('rm /classifier.bin')
  runcmd('rm /classifier.param')
  runcmd('cp /content/v831_yolo/out/yolov2.bin /classifier.bin')
  runcmd('cp /content/v831_yolo/out/yolov2.param /classifier.param')
  runcmd('cp -rf /content/v831_yolo/out/test /images')
  runcmd('zip -r /classifier.zip /images  /classifier.bin /classifier.param')

TEMPLATE_DIR = os.path.abspath(PATH +'/kidbrightMAI')
UPLOAD_FOLDER = os.path.join(PATH, 'uploads')

app = Flask(__name__, template_folder=TEMPLATE_DIR)

app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER

@app.route('/')
def main():
    return render_template("index.html")


@app.route('/uploadIM', methods=['POST'])
def uploadIM():
    if request.method == 'POST':


        try:
          folder_path = '/content/v831_restnet18/data'
          shutil.rmtree(folder_path)
          print('Data Folder and its content removed')
        except:
          print('Folder not deleted')
        if not os.path.exists("/content/v831_restnet18/data"):
          os.makedirs("/content/v831_restnet18/data")

        try:
          folder_path = '/content/v831_restnet18/data/model'
          shutil.rmtree(folder_path)
          print('Data Folder and its content removed')
        except:
          print('Folder not deleted')


        try:
          folder_path = '/content/v831_restnet18/out'
          shutil.rmtree(folder_path)
          print('Out Folder and its content removed')
        except:
          print('Folder not deleted')
        if not os.path.exists("/content/v831_restnet18/out"):
          os.makedirs("/content/v831_restnet18/out")
        try:
          folder_path = '/content/v831_restnet18/images'
          shutil.rmtree(folder_path)
          print('Out Folder and its content removed')
        except:
          print('Folder not deleted')
        if not os.path.exists("/content/v831_restnet18/images"):
            os.makedirs("/content/v831_restnet18/images")

        # Get the list of files from webpage
        files = request.files.getlist("file")

        # Iterate for each file in the files List, and Save them
        for file in files:
            #file.save(file.filename)
            file.save(os.path.join(app.config['UPLOAD_FOLDER'], file.filename))
        print(files[0].filename)
        cmd2 = 'unzip /content/uploads/' + files[0].filename + ' -d /content/v831_restnet18/data'
        proc2 = check_output(cmd2, shell = True)
        cmd3 = 'rm /content/v831_restnet18/data/dataset.json'
        proc3 = check_output(cmd3, shell = True)
        cmd4 = 'rm /content/v831_restnet18/data/project.json'
        proc4 = check_output(cmd4, shell = True)
        cmd5 = 'cp -rf /content/v831_restnet18/data/dataset/* /content/v831_restnet18/data'
        proc5 = check_output(cmd5, shell = True)
        cmd6 = 'rm -rf /content/v831_restnet18/data/dataset'
        proc6 = check_output(cmd6, shell = True)

        outputdir = "/content/v831_restnet18/images"

        classes = os.listdir("/content/v831_restnet18/data")
        dirtest = os.path.join("/content/v831_restnet18/data", classes[0])
        files = [file for file in os.listdir(dirtest) if os.path.isfile(os.path.join(dirtest, file))]
        print(files)

        if len(files) < 15:
            for file in files:
                shutil.copyfile(os.path.join(dirtest, file), os.path.join(outputdir, file))
        else:
            # Amount of random files you'd like to selec
            random_amount = 15
            for x in range(random_amount):
                if len(files) == 0:
                    break
                else:
                    file = random.choice(files)
                    shutil.copyfile(os.path.join(dirtest, file), os.path.join(outputdir, file))


        x = threading.Thread(target=training)
        x.start()



        return "<h1>Files Uploaded Successfully.!</h1>"

@app.route('/upload', methods=['POST'])
def upload():
    if request.method == 'POST':

        try:
          folder_path = '/content/uploads'
          shutil.rmtree(folder_path)
          print('Data Folder and its content removed')
        except:
          print('Folder not deleted')
        if not os.path.exists("/content/uploads"):
          os.makedirs("/content/uploads")

        try:
          folder_path = '/content/v831_yolo/data/custom'
          shutil.rmtree(folder_path)
          print('Data Folder and its content removed')
        except:
          print('Folder not deleted')
        if not os.path.exists("/content/v831_yolo/data/custom"):
          os.makedirs("/content/v831_yolo/data/custom")

        try:
          folder_path = '/content/v831_yolo/out'
          shutil.rmtree(folder_path)
          print('Out Folder and its content removed')
        except:
          print('Folder not deleted')
        if not os.path.exists("/content/v831_yolo/out"):
          os.makedirs("/content/v831_yolo/out")

        try:
          folder_path = '/content/v831_yolo/out/test'
          shutil.rmtree(folder_path)
          print('Out Folder and its content removed')
        except:
          print('Folder not deleted')
        if not os.path.exists("/content/v831_yolo/out/test"):
            os.makedirs("/content/v831_yolo/out/test")


        try:
          folder_path = '/content/v831_yolo/weights/custom/slim_yolo_v2'
          shutil.rmtree(folder_path)
          print('Out Folder and its content removed')
        except:
          print('Folder not deleted')
        if not os.path.exists("/content/v831_yolo/weights/custom/slim_yolo_v2"):
            os.makedirs("/content/v831_yolo/weights/custom/slim_yolo_v2")

        # Get the list of files from webpage
        files = request.files.getlist("file")

        # Iterate for each file in the files List, and Save them
        for file in files:
            #file.save(file.filename)
            file.save(os.path.join(app.config['UPLOAD_FOLDER'], file.filename))
        print(files[0].filename)
        cmd2 = ['unzip', '/content/uploads/' + files[0].filename, '-d', '/content/uploads/']
        proc2 = check_output(cmd2)
        cmd3 = 'cp -rf /content/uploads/dataset/* /content/v831_yolo/data/custom'
        proc3 = check_output(cmd3, shell = True)
        cmd4 = 'mv /content/v831_yolo/data/custom/ImageSets/Main/valid.txt /content/v831_yolo/data/custom/ImageSets/Main/val.txt'
        proc4 = check_output(cmd4, shell = True)

        res = []

        dir_path = '/content/v831_yolo/data/custom/JPEGImages'
        an_path = '/content/v831_yolo/data/custom/Annotations'
        # Iterate directory
        for file_path in os.listdir(dir_path):
            # check if current file_path is a file
            if os.path.isfile(os.path.join(dir_path, file_path)):
                # add filename to list
                res.append(file_path)
        print(res)
        fnames = []
        for x in res:
          fnames.append(os.path.splitext(x)[0])
        print(fnames)

        ii = 1

        for ff in fnames:
          jpgin = os.path.join(dir_path, ff) + '.jpg'
          xmlin = os.path.join(an_path, ff) + '.xml'
          jpgout = os.path.join(dir_path, str(ii)) + '.jpg'
          xmlout = os.path.join(an_path, str(ii)) + '.xml'
          #print(jpgin)
          #print(xmlin)
          #print(jpgout)
          #print(xmlout)
          os.rename(jpgin, jpgout)
          os.rename(xmlin, xmlout)
          ii = ii + 1


        for ii in range(1, len(fnames) + 1):
          #print(ii)
          ffname = '/content/v831_yolo/data/custom/Annotations/' + str(ii)  +'.xml'

          tree = ET.parse(ffname)
          root = tree.getroot()
          # parsing using the string.
          #stringroot = ET.fromstring(XMLexample_stored_in_a_string)
          # printing the root.
          #print(root)

          for filename in root.iter('filename'):
            #print(filename.text)
            filename.text = str(ii) + '.jpg'

          for path in root.iter('path'):
            #print(filename.text)
            path.text = str(ii) + '.jpg'


          tree.write(ffname)


        file1 = open('/content/v831_yolo/data/custom/ImageSets/Main/train.txt', 'w')
        for ii in range(1, len(fnames) - 10 + 1):
          file1.write(str(ii) + '\n')
        file1.close() #close file

        file2 = open('/content/v831_yolo/data/custom/ImageSets/Main/val.txt', 'w')
        for ii in range(1, 11):
          file2.write(str(len(fnames) - 10 + ii) + '\n')
        file2.close() #close file

        file3 = open('/content/v831_yolo/data/custom/ImageSets/Main/valid.txt', 'w')
        for ii in range(1, 11):
          file3.write(str(len(fnames) - 10 + ii) + '\n')
        file3.close() #close file

        x = threading.Thread(target=trainobj)
        x.start()




        return "<h1>Files Uploaded Successfully.!</h1>"

@app.route('/genmod', methods=['POST'])
def genmod():
  if request.method == 'POST':
    x = threading.Thread(target=genmodt)
    x.start()
    return "<h1>Genmod OK.!</h1>"


from google.colab.output import eval_js
print(eval_js("google.colab.kernel.proxyPort(5000)"))
app.run(host='0.0.0.0', port=5000)
```
